#### 01. augmented Dickey-Fuller Test
- 시계열의 정상여부를, 수치를 통해 양적으로 알려주는 방법

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller(df1['Thousands of Passengers'])
```

```python
# help(adfuller)       
```

```python
dftest = adfuller(df1['Thousands of Passengers'])
dfout = pd.Series(dftest[0:4], index=['ADF Test Statistic', 'p-value', '# Lags Used', '# Observations']) # 시차 수, 관찰 수
```

```python
for key, val in dftest[4].items():
    dfout[f'critical value ({key})'] = val
```

```python
# 본 강의에서 디키풀러 테스트를 위해 만들어둔 함수
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
adf_test(df1['Thousands of Passengers'])
```

#### 02. Granger Causality Test
- 시계열의 인과관계 검정

```python
df3 = pd.read_csv('samples.csv', index_col=0, parse_dates=True)
df3.index.freq = 'MS'
```

```python
df3.head()
```

```python
# 눈으로만 확인하면 a와 d에 어떤 상관관계가 있는지 확인하기 어려움
df3[['a', 'd']].plot(figsize=(12,8))
```

```python
# d에 변화가 있고 이틀 후에 a에 그 변화가 반영된 것처럼 보임
df3['a'].iloc[2:].plot(figsize=(12,8), legend=True)
df3['d'].shift(2).plot(legend=True)
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
# maxlang 으로 시차를 얼마나 정할지는 직접 설정해야 하는 문제
# trade-off 관계를 생각해서 설정해야 함
# 확인하는 시차의 범위가 클수록 인과관계를 발견하게 될 가능성은 높아도 시간은 오래걸림
grangercausalitytests(df3[['a','d']], maxlag=3);
```

```python
# 인과관계가 명확해서 p 값이 0.05 보다 작았던 a와 d랑 달리
# 인과관계가 없는 b와 d의 p 값은 0.05 보다 큼
grangercausalitytests(df3[['b','d']], maxlag=3);
```

#### 03. forecast evaluation
- 평균제곱오차(MSE), 평균제곱근오차(RMSE), 평균절대오차(MAE)

```python
np.random.seed(42)

df = pd.DataFrame(np.random.randint(20,30,(50,2)), columns=['test','predictions'])
```

```python
df.head()
```

```python
df.plot(figsize=(12,8))
```

```python
from statsmodels.tools.eval_measures import mse, rmse, meanabs
```

```python
# 어차피 차이값이 양수가 되기 때문에 순서는 상관없음
mse(df['test'], df['predictions'])
```

```python
# mse, mae, rmse 중에 rmse를 보통 평가 기준으로 사용함(=단위보정)
rmse(df['test'], df['predictions'])
```

```python
meanabs(df['test'], df['predictions'])
```
